# How to read "cine_seg.npz"

#### step 1. Load "cine_seg.npz" 

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

path="./cine_seg.npz" # input the path of cine_seg.npz in your environment
data=np.load(path,allow_pickle=True)
```

#### step 2. The names of each object in data can be checked using the files attribute.

```python
print(f"data contains {len(data.files)} objects, and the name of each object is below:")
print(data.files)
```

#### step 3. Each object contains a image and corresponding ground_truth.

The pixel's value in ground_truth is one of [ 0, 85, 170, 255]. '0' is for background, '85' is for right ventricle(RV), '170' is for myocardium(MYO), '255' is for left ventricle(LV)

```python
img=data['patient001_01_2'][0]
gt=data['patient001_01_2'][1]
plt.figure(figsize=(10,4))
plt.subplot(1,2,1)
plt.imshow(img)
plt.colorbar()
plt.subplot(1,2,2)
plt.imshow(gt)
plt.colorbar()
plt.show()
plt.close('all')
```
